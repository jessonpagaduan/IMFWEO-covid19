```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
url1 = 'https://raw.githubusercontent.com/jessonpagaduan/IMFWEO-covid19/master/weo_june2020.csv'
df1 = pd.read_csv(url1, index_col=0)
df1
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 10), sharey=True, sharex=True)

ax1 = df1[['fch_gdp_2020', 'fch_gdp_2021']].plot(kind='barh', ax=axs[0], legend=False)
ax1.invert_yaxis()
ax1.set(title='GDP growth (% change)', xlabel='Percentage points')

ax1 = df1[['fch_fb_2020', 'fch_fb_2021']].plot(kind='barh', ax=axs[1], legend=False)
ax1.invert_yaxis()
ax1.set(title='Fiscal balance (% of GDP)', xlabel='Percentage points')

ax1 = df1[['fch_gd_2020', 'fch_gd_2021']].plot(kind='barh', ax=axs[2], legend=False)
ax1.invert_yaxis()
ax1.set(title='Gross debt (% of GDP)', xlabel='Percentage points')
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 10), sharey=True)

ax1 = df1['fch_gdp_2020'].plot(kind='barh', ax=axs[0], legend=False)
ax1.invert_yaxis()
ax1.set(title='GDP growth (% change)', xlabel='Percentage points')

ax1 = df1['fch_fb_2020'].plot(kind='barh', ax=axs[1], legend=False)
ax1.invert_yaxis()
ax1.set(title='Fiscal balance (% of GDP)', xlabel='Percentage points')

ax1 = df1['fch_gd_2020'].plot(kind='barh', ax=axs[2], legend=False)
ax1.invert_yaxis()
ax1.set(title='Gross debt (% of GDP)', xlabel='Percentage points')

fig.subplots_adjust(wspace=0.1)

plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 10), sharey=True)

ax1 = df1['fch_gdp_2020'].plot(kind='barh', ax=axs[0], legend=False)
ax1.invert_yaxis()
ax1.set(title='GDP growth (% change)', xlabel='Percentage points', xlim=(-7, 7))

ax1 = df1['fch_fb_2020'].plot(kind='barh', ax=axs[1], legend=False)
ax1.invert_yaxis()
ax1.set(title='Fiscal balance (% of GDP)', xlabel='Percentage points', xlim=(-11, 11))

ax1 = df1['fch_gd_2020'].plot(kind='barh', ax=axs[2], legend=False)
ax1.invert_yaxis()
ax1.set(title='Gross debt (% of GDP)', xlabel='Percentage points', xlim=(-16, 16))

fig.subplots_adjust(wspace=0.1)

plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 8), sharey=True)


color1 = []
for country in df1.dropna()['fch_gdp_2020'].index:
    if country == 'Australia':
        color1.append('gray')
    else:
        color1.append('red')
ax1 = df1.dropna()['fch_gdp_2020'].plot(kind='barh', color=color1, ax=axs[0], legend=False)
ax1.invert_yaxis()
ax1.set(xlabel='Percentage points', xlim=(-7, 7))
ax1.set_title('Change in projections of \nGDP growth as % change y-o-y ', fontsize=12)

color2 = []
for country in df1.dropna()['fch_fb_2020'].index:
    if country in ['Australia', 'Saudi Arabia']:
        color2.append('gray')
    else:
        color2.append('red')
ax2 = df1.dropna()['fch_fb_2020'].plot(kind='barh', color=color2, ax=axs[1], legend=False)
ax2.invert_yaxis()
ax2.set(xlabel='Percentage points', xlim=(-9, 9))
ax2.set_title('Change in projections of \nfiscal balance as % of GDP', fontsize=12)

color3 = []
for country in df1.dropna()['fch_gd_2020'].index:
    if country in ['Australia', 'Canada', 'China']:
        color3.append('gray')
    else:
        color3.append('red')
ax3 = df1.dropna()['fch_gd_2020'].plot(kind='barh', color=color3, ax=axs[2], legend=False)
ax3.invert_yaxis()
ax3.set(xlabel='Percentage points', xlim=(-17, 17))
ax3.set_title('Change in projections of \ngross debt as % of GDP', fontsize=12)

fig.suptitle('Prospects for a V-shaped recovery look unlikely as the global economic outlook for 2020 \
appears grim \nwith sizeable downward revisions in GDP growth projections, \
larger expected fiscal deficits, and elevated debt forecasts.', y=1.03, fontsize=15)
fig.subplots_adjust(wspace=0.05)

plt.annotate('Source: IMF World Economic Outlook database (accessed 13 July 2020).', 
             xy=(-2.4, -0.1), xycoords='axes fraction', textcoords='offset points', va='top')
plt.show()
```
